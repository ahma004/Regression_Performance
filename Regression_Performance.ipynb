````python
def Sum_Of_Squares(x, y, hFunc):
    # Starter resultatet på 0
    result = 0

    # Gennemgår x- og y-værdier parvist
    for numX, numY in zip(x, y):
        # Finder forskellen mellem modellens forudsigelse og den rigtige y-værdi
        dif = hFunc(numX) - numY

        # Lægger den kvadrerede fejl til resultatet
        result = result + dif**2

    # Returnerer summen af alle kvadrerede fejl
    return result
````

### Assigment 5: RMSE function

Declare a function for calculating and returning the value of RMSE.
You just need to utilize `Sum_Of_Squares`, a square root and division by the number of data points:

```
def RMSE(x, y, hFunc):  
    result=0
    MSE = 0
    n=len(x)
    if n > 0:
      MSE= Sum_Of_Squares(x, y, hFunc)/n

    return math.sqrt(MSE)
```
Copy the definition to a code cell below and add comments (#....) to it on what it is doing step by step.

Print out the error for different values of a and b.


````python
def RMSE(x, y, hFunc):
    # Starter med at sætte result og MSE til 0
    result = 0
    MSE = 0

    # Finder antal datapunkter
    n = len(x)

    # Hvis der er datapunkter
    if n > 0:
        # Beregner Mean Squared Error som sum of squares divideret med n
        MSE = Sum_Of_Squares(x, y, hFunc) / n

    # Returnerer kvadratroden af MSE = RMSE
    return math.sqrt(MSE)
````

````python
print(RMSE(x, y, h))
````

### Assignment 7: Discussion in the class
So what is best linear or polynomial regression ?
Can we conclude?
Lineær regression er bedst, når data følger en nogenlunde ret linje. Modellen er enkel, let at forstå og hurtig at arbejde med.

Polynomial regression er bedre, når data har en buet udvikling. Den kan tilpasse sig mere komplekse mønstre og vil ofte få en lavere RMSE, hvis datasættet ikke er lineært.

Man kan dog ikke bare sige, at polynomial regression altid er bedst. Hvis modellen bliver for kompleks, kan den overfitte data. Derfor skal man både se på RMSE og på, om modellen giver mening i forhold til datasættet.

Konklusion: Den model med lavest RMSE passer bedst til netop dette datasæt, men den simpleste model kan stadig være at foretrække, hvis forskellen i fejl ikke er stor.
